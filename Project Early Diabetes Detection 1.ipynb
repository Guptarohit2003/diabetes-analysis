```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('diabetes_data_upload.csv')
```

```python
df = df.replace("No",0)
df = df.replace("Yes",1)

df = df.replace("Positive",1)
df = df.replace("Negative",0)

#let gender == ismale
df = df.replace("Male" ,1)
df = df.replace("Female", 0)
```

```python
replace = {"Gender" : "isMale"}
df = df.rename(columns = replace)
```
